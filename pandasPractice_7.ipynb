```python
import pandas as pd
```

```python
sales = pd.DataFrame({
    'Store': ['A', 'A', 'B', 'B', 'C', 'C', 'C'],
    'Item': ['Pen', 'Notebook', 'Pen', 'Notebook', 'Pen', 'Pen', 'Notebook'],
    'Quantity': [10, 5, 7, 8, 12, 15, 9],
    'Price': [20, 50, 22, 55, 19, 21, 49],
    'Date': pd.to_datetime(['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-04'])
})
```

```python
sales.groupby('Item')['Price'].sum().reset_index(name='Total_revenue')
```

```python
i = sales.groupby(['Store'])['Price'].idxmax()
sales.iloc[i]
```

```python
agg_df = sales.groupby('Item').agg({'Price': ['mean', 'sum'], 'Quantity': 'sum'})
agg_df['price_per_quantity'] = agg_df[('Price', 'sum')] / agg_df[('Quantity', 'sum')]
agg_df.reset_index()
```

```python
students = pd.DataFrame({
    'Student': ['Ali', 'Neha', 'Tom', 'Jaya', 'Ali', 'Tom', 'Neha'],
    'Subject': ['Math', 'Math', 'Math', 'Math', 'Physics', 'Physics', 'Physics'],
    'Score': [88, 92, 78, 85, 91, 80, 89],
    'Grade_Level': [10, 10, 10, 10, 11, 11, 11]
})
```

```python
students.groupby(['Subject','Grade_Level'])['Score'].mean().reset_index(name='Avg_score')
```

```python
x = students.groupby('Subject')['Score'].var().reset_index(name='Score_variation')
x[x.Score_variation == x.Score_variation.max()]
```

```python
students.assign(rank = students.groupby('Student')['Score'].rank(method='dense',ascending=False))
```

```python
students.pivot_table(index='Subject',columns='Grade_Level',aggfunc=({'Score':'mean'})).reset_index()
```

```python
experiments = pd.DataFrame({
    'Test_ID': ['T1', 'T2', 'T3', 'T4', 'T5', 'T6'],
    'Material': ['Steel', 'Steel', 'Copper', 'Copper', 'Copper', 'Steel'],
    'Batch': ['B1', 'B1', 'B2', 'B2', 'B3', 'B1'],
    'Strength': [500, 520, 460, 470, 480, 510],
    'Elasticity': [0.8, 0.82, 0.75, 0.76, 0.78, 0.81]
})
```

```python
experiments.groupby('Material').agg({'Strength':['mean','std']}).reset_index()
```

```python
experiments.assign(rank = experiments.groupby('Batch')['Elasticity'].transform('mean').rank(method='dense'))
```

```python
experiments.groupby(['Material','Batch']).describe().reset_index()
```

```python
experiments.groupby('Batch').agg({'Strength':['min','max',lambda x:x.max()-x.min()]})
```
