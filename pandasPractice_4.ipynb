```python
def check(row):
  fname = row['Fname']
  fph = row['Fph']
  mname = row['Mname']
  mph = row['Mph']
  if (fname is not None) and (fname in names) and (fph not in nums):
    return True
  elif (fname is not None) and (mname in names) and (mph not in nums):
    return True
  else:
    return False
std_copy['Father_mismatch'] = std_copy.apply(check,axis=1)
std_copy['Mother_mismatch'] = std_copy.apply(check,axis=1)
```
