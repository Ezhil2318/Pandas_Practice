```python
import pandas as pd

df_raw = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South'],
    'Product': ['Apples', 'Oranges', 'Apples', 'Oranges'],
    'Sales': [100, 150, 200, 250]
})
```

```python
df_pivot = df_raw.pivot(index='Region', columns='Product', values='Sales')
df_pivot
```

```python
df_melted = df_pivot.reset_index().melt(id_vars='Region', var_name='Product', value_name='Sales')
df_melted
```

```python
df_raw.equals(df_melted)
```

```python
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['A', 'B']})
df2 = pd.DataFrame({'ID': [1, 2], 'Score': [90, 80]})

# Using merge (recommended)
pd.merge(df1, df2, on='ID')
```

```python
# Using join (requires setting index)
df1.set_index('ID').join(df2.set_index('ID'))
```

```python
df = pd.DataFrame({'Group': ['B', 'A', 'C'], 'Value': [10, 20, 30]})

# Default sorts keys
df.groupby('Group').sum()
```

```python
# Disable sorting for speed
df.groupby('Group', sort=False).sum()
```

```python
df_missing = pd.DataFrame({
    'Region': ['North', 'South', 'South'],
    'Product': ['Apples', 'Apples', 'Oranges'],
    'Sales': [100, 200, None]
})

df_pivoted = df_missing.pivot(index='Region', columns='Product', values='Sales')
df_unstacked = df_pivoted.stack(dropna=False).fillna(0)
```

```python
df = pd.DataFrame({
    'user_id': [101, 101, 102, 102],
    'channel': ['web', 'mobile', 'web', 'mobile'],
    'spend': [500, 200, 600, 100]
})

# Step-by-step process
df_filtered = df[df['spend'] > 100]
df_grouped = df_filtered.groupby(['user_id', 'channel'])['spend'].sum()
df_pivoted = df_grouped.unstack(fill_value=0)
```
