```python
import pandas as pd
from datetime import datetime, timedelta


employees_df = pd.DataFrame({
    'emp_id': [101, 102, 103, 104, 105, 106],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'dept_id': [1, 2, 1, 3, 2, 3],
    'salary': [55000, 62000, 52000, 58000, 60000, 54000],
    'join_date': pd.to_datetime(['2021-01-10', '2020-06-23', '2021-03-12', '2020-09-01', '2024-07-15', '2022-03-03']),
    'manager_id': [None, 101, 101, 102, 102, 104]
})

departments_df = pd.DataFrame({
    'dept_id': [1, 2, 3, 4],
    'dept_name': ['HR', 'Finance', 'IT', 'Marketing'],
    'location': ['Mumbai', 'Delhi', 'Chennai', 'Bangalore']
})

projects_df = pd.DataFrame({
    'proj_id': [301, 302, 303, 304, 305, 306],
    'proj_name': ['Onboarding App', 'Budget Tracker', 'Compliance Portal', 'Website Revamp', 'Forecast Tool', 'Audit Automation'],
    'emp_id': [101, 102, 103, 104, 105, 106],
    'dept_id': [1, 2, 1, 3, 2, 3],
    'hours_worked': [120, 95, 110, 150, 85, 130],
    'deadline': pd.to_datetime(['2025-08-01', '2025-11-15', '2025-10-30', '2025-08-15', '2025-09-05', '2025-12-20'])
})
```

```python
x = employees_df.merge(departments_df,on='dept_id',how='inner')
x = x.merge(projects_df,on='emp_id',how='inner',suffixes=("","_proj"))
x[(x.dept_name.str.startswith('F'))|(x.dept_name.str.startswith('I'))]
```

```python
today = pd.Timestamp('today').normalize()
n30 = today + pd.Timedelta(days=30)
n60 = today + pd.Timedelta(days=60)
deadline_bucket = pd.Series('Future',index=projects_df.index)
deadline_bucket = deadline_bucket.where(~projects_df.deadline.between(n30,n60),'Later')
deadline_bucket =  deadline_bucket.where(~(projects_df.deadline<n30),'Soon')
projects_df.assign(deadline_bucket = deadline_bucket)
```

```python
def bucket(deadline):
  if deadline<n30:
    return 'Soon'
  elif deadline<n60:
    return 'Later'
  else:
    return 'Future'
projects_df.assign(deadline_bucket = projects_df['deadline'].apply(bucket))
```

```python
tenure = pd.Series('Veteran',index=employees_df.index)
tenure = tenure.where(~((today.year-employees_df['join_date'].dt.year).between(2,4)),'Mid')
tenure = tenure.where(~(today.year-employees_df['join_date'].dt.year<2),'New')
employees_df.assign(tenure = tenure)
```
