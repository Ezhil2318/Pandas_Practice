```python
Map = telephone_df.dropna(subset='Name').set_index('Name')['PH_num']
condition1 = students_df.Fph.isna()&students_df.Fname.notna()
condition2 = students_df.Mph.isna()&students_df.Mname.notna()
students_df.loc[condition1,'Fph'] = students_df.loc[condition1,'Fname'].map(Map)
students_df.loc[condition2,'Mph'] = students_df.loc[condition2,'Mname'].map(Map)
students_df
```

```python
'''
names  = list(telephone_df['Name'].dropna())
students_df['listed'] = (students_df.Fname.isin(names)|students_df.Mname.isin(names))
students_df
'''

names  = list(telephone_df['Name'].dropna())
def check(row):
  Fname=row['Fname']
  Mname=row['Mname']
  if (Fname in names) and (Mname in names):
    return "Both"
  elif (Fname in names) or (Mname in names):
    return "Either"
  else:
    return "Neither"
students_df['listed'] = students_df.apply(check,axis=1)
students_df.groupby('listed')['Std_name'].size().reset_index(name='Count')
```

```python
fcity = (students_df.merge(telephone_df,left_on='Fph',right_on='PH_num',how='inner').dropna(subset=['Name']).drop_duplicates(subset='Name').set_index('Name')['City'])
mcity = (students_df.merge(telephone_df,left_on='Mph',right_on='PH_num',how='inner').dropna(subset=['Name']).drop_duplicates(subset='Name').set_index('Name')['City'])
students_df['City'] = students_df['Fname'].map(fcity)
students_df['City'] = students_df['City'].fillna(students_df['Mname'].map(mcity))
students_df
```

```python
phnum = list(telephone_df.PH_num)
condition1 = students_df.Fname.notna()|students_df.Mname.notna()
condition2 = (students_df.Fph.isin(phnum)==False)|(students_df.Mph.isin(phnum)==False)
students_df[condition1&condition2]
```

```python
fstate = (students_df.merge(telephone_df,left_on='Fph',right_on='PH_num',how='inner')).dropna(subset=['Name']).drop_duplicates(subset='Name').set_index('Name')['State']
mstate = (students_df.merge(telephone_df,left_on='Mph',right_on='PH_num',how='inner')).dropna(subset=['Name']).drop_duplicates(subset='Name').set_index('Name')['State']
def states(row):
  fname = row['Fname']
  mname = row['Mname']
  if mname in mstate:
    return mstate[mname]
  elif fname in fstate:
    return fstate[fname]
  else:
    return None
students_df['Home_state'] = students_df.apply(states,axis=1)
```

```python
def whose(row):
  fname = row['Fname']
  mname = row['Mname']
  if (fname in names) and (mname not in names):
    return 'Father'
  elif (fname not in names) and (mname in names):
    return 'Mother'
  elif (fname in names) and (mname in names):
    return 'Both'
  else:
    return None
students_df['Whose_listed'] = students_df.apply(whose,axis=1)
students_df[students_df['Whose_listed']!='Both']
```
